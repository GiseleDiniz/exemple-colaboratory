```python
covid = response.json()

titles = [item for item in covid['hits']['hits'][2]['_source']]
values = [item for item in covid['hits']['hits'][2]['_source'].values()]


dicionario = {"title": titles, "values": values}

pd.DataFrame(dicionario, columns=['title', 'values'])
```

```python
#pegar a coluna da df
lista_arrumar = list([item['_source']['testes'] for item in covid['hits']['hits']])
lista_geral = list()
for i in lista_arrumar:
  if len(i) > 1:
    lista_resultado = list()
    for j in i:
      teste = pd.DataFrame.from_dict([j])["resultadoTeste"][0]
      lista_resultado.append(teste)
    lista_geral.append(lista_resultado)
  elif len(i) == 1:
    for j in i:
      string_aux = pd.DataFrame.from_dict([j])["resultadoTeste"][0]
      lista_geral.append(string_aux)
  else:
    lista_geral.append("NA")
df_covid["resultadoTeste"] = lista_geral
```

```python
#excluir a coluna
df_covid.drop(['_source.testes'], axis=1)
```

```python
#Qual o estado com a maior proporção de casos confirmados?
df_reagente = df_covid.loc[df_covid['resultadoTeste']== "Reagente"].groupby(['_source.estado'])['resultadoTeste'].count().sort_values().reset_index()
print(df_reagente)
```

O estado com maior número de casos confirmados é o Rio Grande do Sul, contendo 34 casos confirmados

```python
#Dentre os casos confirmados, qual a proporção de pacientes que receberam ao menos 1 dose de vacina?
def vacina():
  indicador1= df_covid.loc[df_covid['_source.dataPrimeiraDose'] != "None"]
  indicadores2= indicador1.loc[df_covid['resultadoTeste']== "Reagente"]
  valores = indicadores2.groupby(['resultadoTeste'])['_source.dataPrimeiraDose'].count().reset_index()
  
  return valores
```

```python
#Como é a distribuição de idade entre pacientes sintomáticos e assintomáticos entre os casos confirmados?
def sintomas():
  indicador1= df_covid.loc[df_covid['_source.sintomas'] != "None"]
  indicadores2= indicador1.loc[df_covid['resultadoTeste']== "Reagente"].groupby(['_source.idade',"_source.sintomas"])['resultadoTeste'].count().reset_index()
  #valores = indicadores2.groupby(['resultadoTeste', '_source.sintomas'])['_source.idade'].count().reset_index()
  return indicadores2
```

```python
#Para (pelo menos um) estados a sua escolha, construa uma visualização para acompanhar a evolução dos casos ao longo do período amostrado.
def acompanhamento():
  indicador1= df_covid.loc[df_covid['_source.estado'] == "Pernambuco"]
  indicador2= indicador1.loc[df_covid['resultadoTeste'] == "Reagente"]
  indicador3 = indicador2.loc[df_covid['_source.evolucaoCaso'] != "None"]
  Valormed=indicador3.groupby(['_source.evolucaoCaso', '_source.estado'])['resultadoTeste', '_source.dataNotificacao'].count().reset_index()
  Valormed=Valormed.sort_values(by=["_source.estado","_source.evolucaoCaso"])
  return Valormed
fig = px.bar(acompanhamento(), x="_source.evolucaoCaso", y="resultadoTeste", title='Evolucao dos casos no estado do Rio Grande do Sul')
fig.show()
```
